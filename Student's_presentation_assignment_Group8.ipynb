The **exponential distribution** is a continuous probability distribution that is widely used to model the time between events in a Poisson process. It is particularly useful for modeling scenarios where events occur continuously and independently at a constant average rate.

**Explanation of Using the CDF to Calculate the Probability**


The question asks for the probability that the next customer arrives within 10 minutes. This is a typical problem where the cumulative distribution function (CDF) of an exponential distribution is used.

Why use the CDF?

The cumulative distribution function (CDF) gives the probability that a random variable

X is less than or equal to a certain value

x.
In this context, the random variable

X represents the time between customer arrivals.
The CDF for an exponential distribution with rate parameter

位 (which is 3 customers per hour in this case) is given by: F(x)=1e^
位x

**Numpy implementation of the Example**

```python
customer_rate=3
time_minuets=10

#converting time in minutes to Hours
time_hours = time_minuets/60

# calculate the probability using exponential CDf

probability= 1-np.exp(-customer_rate*time_hours)

print(f'The probabity that the next customer will arrive within 10 minutes, f(x<1/6): {probability: .4f}')
```

**We can plot the PDF and CDF distributions using Mathplotlib just to visualize the data**: This is optional and not related to the assignment

```python
import matplotlib.pyplot as plt
```

```python
# Generating the plotting data
time_values = np.linspace(0, 2, 500)
pdf_values = customer_rate * np.exp(-customer_rate * time_values)
cdf_values = 1 - np.exp(-customer_rate * time_values)

# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plotting PDF
ax1.plot(time_values, pdf_values, label="PDF", color='blue')
ax1.set_title('Probability Density Function')
ax1.set_xlabel('Time (Hours)')
ax1.set_ylabel('Density')
ax1.axvline(x=time_hours, color='r', linestyle='--', label='10 minutes')
ax1.legend()

# Plotting CDF
ax2.plot(time_values, cdf_values, label="CDF", color='green')
ax2.set_title('Cumulative Distribution Function')
ax2.set_xlabel('Time (Hours)')
ax2.set_ylabel('Cumulative Probability')
ax2.axvline(x=time_hours, color='r', linestyle='--', label='10 minutes')
ax2.axhline(y=probability, color='r', linestyle='--')
ax2.legend()

plt.suptitle('Customer Arrival Distribution in a Coffee Shop')
plt.show()
```

**Ingterpretation of the graphs**

PDF: The blue curve shows that the likelihood of customer arrivals is highest immediately after the previous customer (time close to 0) and decreases as time passes. The red line at 10 minutes shows that there's a significant likelihood of arrivals before this time.


CDF: The green curve indicates that there is about a 39.35% chance (approximately) that the next customer will arrive within 10 minutes. This is visually shown by the red dashed lines marking 10 minutes on the x-axis and the corresponding cumulative probability on the y-axis.
